# Music Classification using Deep Learning 

```python
import pandas as pd
import numpy as np

pd.set_option('display.max_rows', 500)

import keras
from keras.layers import Dense, Dropout
from keras.models import Sequential

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder,MinMaxScaler
from keras.utils import np_utils
from sklearn.metrics import confusion_matrix, accuracy_score
```

```python
df = pd.read_csv("songs_data.csv",index_col=0)
```

```python
df.head(3)
```

```python
#keep only necessary columns
df.drop(['album','artist','id','popularity','release_date','time_signature','key'],inplace=True,axis=1)
```

```python
df.head(2)
```

```python
X=df.iloc[:,:-1]
```

```python
X.head(3)
```

```python
Y=df['emotion']
```

```python
encoder = LabelEncoder()
Y = encoder.fit_transform(Y)
```

```python
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size=0.2,random_state=15)
pd.Series(Y_train).value_counts()
```

```python
ms= MinMaxScaler()
X_tr = ms.fit_transform(X_train.iloc[:,1:])
```

```python
X_tr.shape
```

```python
dummy_y_tr = np_utils.to_categorical(Y_train)
dummy_y_tr.shape
```

```python
model = Sequential()
model.add(Dense(64,input_dim=10,activation='relu'))
model.add(Dropout(0.3))
model.add(Dense(32,activation='relu'))
model.add(Dropout(0.3))
model.add(Dense(4,activation='softmax'))
model.compile(loss='categorical_crossentropy',optimizer='adam',metrics=['accuracy'])
```

```python
model.fit(X_tr,dummy_y_tr, epochs=180, batch_size= 32)
```

```python
X_tst = ms.transform(X_test.iloc[:,1:])
y_pred=model.predict_classes(X_tst)
```

```python
y_pred
```

```python
Y_test
```

```python
accuracy_score(Y_test,y_pred)
```

```python
X_test['result']= y_pred
```

```python
mapping={0:'calm',1:'happy',2:'neutral',3:'sad'}
X_test['result']= X_test.result.map(mapping)
```
